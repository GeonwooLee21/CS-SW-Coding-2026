```python
# [코드 5-10] : 2차 방정식의 근을 구하는 기능을 함수로 만들기
def print_root(a, b, c):
    r1 = (-b + (b**2 - 4*a*c)**0.5) / (2*a)
    r2 = (-b - (b**2 - 4*a*c)**0.5) / (2*a)
    print('해는', r1, '또는', r2)

print_root(1, 2, -8)
print_root(2, -6, -8)
```

```python
# 1. 원의 면적과 둘레를 구하기 위하여 원의 반지름을 사용하려고 한다.
#    이때 사용할 circle_area_circum(radius) 함수는 다음과 같이 반지름(radius) 10을 입력으로 받아서
#    면적(area)과 둘레(circum)의 두 값을 반환하도록 구현하여라.
def circle_area_circum(radius):
    area  = radius ** 2 * 3.14
    circum = radius * 2 * 3.14
    return area, circum

radius = 10
area, circum = circle_area_circum(radius)
print(f'반지름 {radius}인 원의 면적은 {area:.1f}, 원의 둘레는 {circum:.1f}')
```

```python
# 1. 어떤 정수 n과 m을 입력하면, n의 배수 m개를 반환하는 multiples(n, m) 함수를 구현하고
#    다음과 같이 호출하여 결과를 출력하여라.
def multiples(n, m):
    result = []
    for i in range (1, m+1):
        result += [n*i]
    return result

r1, r2, r3, r4 = multiples(3, 4)
print(r1, r2, r3, r4)

r1, r2, r3, r4, r5 = multiples(2, 5)
print(r1, r2, r3, r4, r5)
```

```python
# 1. 가변 인자를 사용하는 sum_nums() 함수를 수정하여 인자들을 튜플 형식으로 출력한 후 
#    모든 값들의 합과 평균을 다음과 같이 출력하시오.
def sum_nums(*nums):
    total = 0;
    for num in nums:
        total += num
        
    average = total / len(nums)
    
    print(f'{len(nums)}개의 인자 {nums}')
          
    return total, average

total, average = sum_nums(10, 20, 30)
print(f'합계 : {total}, 평균 : {average:.1f}')

total, average = sum_nums(10, 20, 30, 40, 50)
print(f'합계 : {total}, 평균 : {average:.1f}')
```
