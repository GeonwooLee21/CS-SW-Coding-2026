## Exercise 3.10

```python
# 3.10 사용자로부터 2개의 정수 a, b를 차례로 입력받은 다음 a가 b의 배수인지 아닌지를 판단하여 출력하는 프로그램을 작성하시오.
a, b = input('두 정수를 입력하시오 : ').split()
a = int(a)
b = int(b)

if b == 0:
    print('0으로 나눌 수 없습니다.')
elif a % b == 0:
    print(f'{a}는(은) {b}의 배수입니다.')
else:
    print(f'{a}는(은) {b}의 배수가 아닙니다.')
```

## Exercise 3.11

```python
# 3.11 1에서 10 사이의 복권 번호 중에서 3개의 숫자를 모두 맞히면 '1억 원', 2개를 맞히면 '1천만 원', 
#      1개를 맞히면 '1만원', 모두 틀리면 '다음 기회에...'를 출력하는 복권 시스템이 있다고 가정하자.
#      사용자로부터 3개의 정수를 받은 다음 상금을 알려주는 프로그램을 다음과 같이 작성하시오.
#      (단, 이번 회차 복권 당첨 번호가 2, 3, 9라고 가정하자.)
num_1, num_2, num_3 = input('세 복권번호를 입력하시오 :').split()
num_1 = int(num_1)
num_2 = int(num_2)
num_3 = int(num_3)

number_of_hits = 0
if (num_1 == 2 or num_1 == 3 or num_1 == 9):
    number_of_hits += 1
if (num_2 == 2 or num_2 == 3 or num_2 == 9) and (num_2 != num_1):
    number_of_hits += 1
if (num_3 == 2 or num_3 == 3 or num_3 == 9) and (num_3 != num_1 and num_3 != num_2):
    number_of_hits += 1

if  number_of_hits == 3:
    print('상금 1억 원')
elif number_of_hits == 2:
    print('상금 1천만 원')
elif number_of_hits == 1:
    print('상금 1만 원')
elif number_of_hits == 0:
    print('다음 기회에...')
```

## Exercise 3.12

```python
# 3.12 중심이 원점 (0,0)에 있고 반지름이 5인 원이 있다고 가정하자. 사용자로부터 x와 y 좌표를 입력 받은 후, 
#      입력받은 점의 좌표 (x, y)가 원의 내부에 있으면 '원의 내부에 있음.', 
#      원의 외부에 있으면 '원의 외부에 있음'을 출력하는 프로그램을 작성하시오.
x, y = input('점의 좌표 x, y를 입력하시오 :').split()
x = int(x)
y = int(y)
distance_from_origin = ((x**2)+(y**2))**0.5

radius = 5

if distance_from_origin <= radius:
    print('원의 내부에 있음')
else:
    print('원의 외부에 있음')
```

## Exercise 4.6

```python
# 4.6 이중 for문을 사용하여 숫자를 입력 받아 다음과 같은 삼각형을 출력하는 프로그램을 작성하시오.
#     이때 다음 결과와 같이 숫자 5가 입력되면 높이가 5이고 제일 마지막 줄의 별 5개가 삼각형 형태로 나타나도록 하시오.
n = int(input('숫자를 입력하세요 : '))

for i in range(1, n + 1):
    line = ''
    for j in range(n - i):
        line += ' '
    for j in range(i):
        line += '*'
    print(line)
```

## Exercise 4.12

```python
# 4.12 뱀 행렬 문제: 사용자로부터 숫자 1보다 크고 10보다 작은 값 n을 입력으로 받아서
#      뱀의 몸통처럼 'ㄹ'과 같은 패턴으로 수가 1씩 증가하는 이차원 배열을 출력하여라.
n = int(input('n을 입력하시오 : '))

for i in range(n):
    line = ''
    start = i * n + 1
    if i % 2 == 0:
        for j in range(n):
            value = start + j
            line += str(value).rjust(3)
    else:
        for j in range(n):
            value = start + n - 1 - j
            line += str(value).rjust(3)
    print(line)
```

## Exercise 4.13

```python
# 4.13 암스트롱 수란 세 자리의 정수 중에서 각 자리의 수를 세제곱한 수의 합과 자신이 같은 수를 말한다.
#      구체적인 예를 들면 153은 1 + 125 + 27의 합으로 구성될 수 있는데 이러한 수가 암스트롱 수이다.
#      세 자리 정수들 중에서 모든 암스트롱 수를 구하여 다음과 같이 출력하여라.
result = '세 자리의 암스트롱 수 : '

for num in range(100, 1000):
    a = num // 100
    b = (num % 100) // 10
    c = num % 10
    
    if num == a**3 + b**3 + c**3:
        result += str(num) + ' '

print(result)
```

## Exercise 4.14

```python
# 4.14 거꾸로 정수는 121이나 3443와 같이 거꾸로 나열해도 그 값이 원래의 값과 같은 정수를 말한다.
#      사용자로부터 숫자 n을 입력 받아 이 수가 거꾸로 정수인지 아닌지를 판단하는 다음과 같은 프로그램을 작성하시오.
n = int(input('정수를 입력하시오 : '))

original = n
reversed_num = 0

while n > 0:
    digit = n % 10
    reversed_num = reversed_num * 10 + digit
    n = n // 10

if original == reversed_num:
    print(f'{original}은(는) 거꾸로 정수입니다.')
else:
    print(f'{original}은(는) 거꾸로 정수가 아닙니다.')
```
