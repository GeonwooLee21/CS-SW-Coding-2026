```python
# 1. 사용자로부터 자동차의 속도(speed)를 km/h 단위의 정수로 입력받도록 하자.
#    자동차의 속도가 100km/h 이상이면 '고속', 100km/h 미만 60km/h 이상이면 '중속', 60km/h 미만이면 '저속'을 출력하는 프로그램을 
#    if-elif-else 문을 이용하여 작성하여라
speed = int(input('자동차의 속도를 입력하세요(단위 : km/h):'))
if  speed >= 100:
    print('고속')
elif speed < 100 and speed >= 60:
    print('중속')
else:
    print('저속')
```

```python
# 1. 미세먼지 농도 예보 기준이 다음과 같을 때 미세먼지 농도를 정수 값으로 입력 받아 '좋음', '보통', '나쁨', '매우 나쁨'을 출력하는 코드를 작성하여라
PM_concentration = int(input('미세먼지의 농도를(단위 : microgram/m^3):'))
if  PM_concentration >= 76:
    print('매우 나쁨')
elif PM_concentration < 76 and PM_concentration >= 36:
    print('나쁨')
elif PM_concentration < 36 and PM_concentration >= 16:
    print('보통')
else:
    print('좋음')
```

```python
# 1. 앞서 배운 누적 덧셈을 응용하여 1에서 100까지 정수의 합을 구하여 출력하여라
# (Hint : range()의 구간이 1에서 100까지가 되도록 하자).
s = 0
for i in range(1, 101):
    s += i
print(s)
```

```python
# 2. range() 함수의 step 값을 이용하여 1에서 100까지 정수 중에서 짝수의 합을 구하여 출력하여라
# (Hint : range() 함수의 시작 값은 0으로 하고 step 값을 2로 하여라).
even_sum = 0
for i in range(0, 101, 2):
    even_sum += i
print(even_sum)
```

```python
# 3. range() 함수의 step 값을 이용하여 1에서 100까지 정수 중에서 홀수의 합을 구하여 출력하여라
# (Hint : range() 함수의 시작 값은 1으로 하고 step 값을 2로 하여라).
odd_sum = 0
for i in range(1, 101, 2):
    odd_sum += i
print(odd_sum)
```

```python
# 1. 이중 for 루프를 이용하여 다음과 같은 패턴을 출력하여라.
# sol_1)
for i in range(1, 8):
    st = ''
    for j in range(1, 9-i):
        if j != 8-i:
            st += ' '
        else:
            st += '#'
    print(st)
```
