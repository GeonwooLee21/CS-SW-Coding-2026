## Exercise 5.9

```python
# 사용자로부터 임의의 수를 연속적으로 입력받도록 하시오. 이 수들에 대한 평균값, 최댓값, 최솟값을 반환하는 함수
# mean_of_n(nums), max_of_n(nums), min_of_n(nums)를 구현하여 다음과 같이 출력하여라.
def mean_of_n(nums):
    return sum(nums)/len(nums)
    
def max_of_n(nums):
    return max(nums)

def min_of_n(nums):
    return min(nums)

inputStr = input('정수를 여러 개 입력하시오 :').split() # split()은 문자열을 특정 구분자(separator)를 기준으로 쪼개어 리스트로 반환하는 메서드
nums = []
for x in inputStr:
    nums += [int(x)]

print(f'평균값은 {mean_of_n(nums):.1f}')
print(f'최댓값은 {max_of_n(nums)}')
print(f'최솟값은 {min_of_n(nums)}')
```

## Exercise 5.16

```python
# 쉼표 단위로 구분된 정수를 임의의 수만큼 입력받은 후 이 숫자들을 오름차순으로 정렬하여 출력하시오.
def ascending_sort(inputList):
    nums = []

    # 문자열 → 정수 변환 후 append
    for x in inputList:
        nums.append(int(x))

    # 입력 리스트 출력
    print(f'입력된 정수의 리스트: {nums}')

    # 정렬
    nums.sort()

    # 정렬 결과 출력
    print('정렬된 정수의 리스트:', end = ' ')
    for n in nums:
        print(n, end = ' ')
    print()

# 입력 처리
inputStr = input('쉼표로 구분된 정수를 여러 개 입력하시오:')
inputList = inputStr.split(',')

# 함수 호출
ascending_sort(inputList)
```

```python
# 두 개의 정수 a, b를 입력으로 받는다. 이때 a < b라는 조건이 항상 지켜진다고 하자.
# 여러분이 작성할 코드는 a에서 b까지(a, b를 포함한다) 범위의 모든 정수의 배수이면서 가장 작은 값을 갖는 값을 찾는 것이다.
# 예를 들어, 2, 4가 입력되면 2, 3, 4 모두의 배수이면서 가장 작은 값이 되는 12가 답이 된다.
# 12는 2의 배수, 3의 배수, 4의 배수이다.
import math

def find_lcm(a, b): # LSM : Least Common Multiple
    lcm = a
    for i in range (a+1, b+1):
        lcm = (lcm * i) // math.gcd(lcm, i)
    return lcm 
        
a = int(input('범위의 시작 정수 :'))
b = int(input('범위의 마지막 정수 :'))
print(f'{a}에서 {b}까지의 정수들의 최소 공배수는 : {find_lcm(a, b)}')
```

##### 주민등록번호의 앞 6자리 생년월일은 두 자리의 년도, 두 자리의 월, 두 자리의 날짜로 표현된다. 예를 들어, 921030은 1992년 10월 30일을 의미한다. 이렇게 주민등록번호 형식의 생년월일이 입력될 때 "1992년 10월 30일"과 같은 형태로 날짜를 출력하는 코드를 작성하라. 입력되는 날짜의 연도는 1950년도 이후부터 입력된다. 즉, 511010은 1951년 10월 10월이지만, 491010은 2049년 10월 10일로 판단해야 한다.

```python
# 1st Approach
def Date_of_Birth(ID_Number):
    day = ID_Number % 100
    
    ID_Number = ID_Number // 100    
    month = ID_Number % 100
    
    ID_Number = ID_Number // 100
    year = ID_Number

    if year >= 50:
        print(f'19{year}년 {month}월 {day}일')
    elif year >= 10:
        print(f'20{year}년 {month}월 {day}일')
    else:
        print(f'200{year}년 {month}월 {day}일')
    

ID_Number = int(input('주민등록번호 첫 6숫자 형식 입력:'))
Date_of_Birth(ID_Number)
```

```python
# 1st Approach (개선)
def date_of_birth(ID_Number): # 파이썬은 보통 함수 이름을 소문자 + 언더바 조합으로 작성함
    day = ID_Number % 100
    
    ID_Number = ID_Number // 100
    month = ID_Number % 100
    
    ID_Number = ID_Number // 100
    year = ID_Number

    if year >= 50:
        print(f'19{year}년 {month}월 {day}일')
    else:
        print(f'20{year:02d}년 {month}월 {day}일') # f-string formatting을 통해 코드를 더 단순하게 만들 수 있음
    

ID_Number = int(input('주민등록번호 첫 6숫자 형식 입력:'))
date_of_birth(ID_Number)
```

```python
# 2nd Approach : 문자열 슬라이싱
def date_of_birth(ID_Number): # 파이썬은 보통 함수 이름을 소문자 + 언더바 조합으로 작성함
    YY = ID_Number[0:2]
    MM = ID_Number[2:4]
    DD = ID_Number[4:6]

    if int(YY) >= 50:
        print(f'19{YY}년 {MM}월 {DD}일')
    else:
        print(f'20{YY}년 {MM}월 {DD}일')
    

ID_Number = input('주민등록번호 첫 6숫자 형식 입력:')
date_of_birth(ID_Number)
```
