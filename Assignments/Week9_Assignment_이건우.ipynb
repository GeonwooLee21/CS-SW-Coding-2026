```python
# 1. datetime 모듈을 사용하여 오늘의 날짜와 시간을 다음과 같이 출력하여라. 
# 이때 hour 값이 10일 경우 오전 10시, 13이면 오후 1시와 같이 오전/오후 정보를 출력하여라.

from datetime import datetime
today = datetime.today()
print(f"오늘의 날짜 : {today.year}년 {today.month}월 {today.day}일")
if today.hour >= 12:
    hour_12 = today.hour - 12 if today.hour > 12 else 12
    print(f"현재사간 : 오후 {hour_12}시 {today.minute}분 {today.second}초")
else:
    hour_12 = today.hour if today.hour != 0 else 12  # 오전 0시가 아니라 오전 12시이므로
    print(f"현재사간 : 오전 {hour_12}시 {today.minute}분 {today.second}초")
```

#### 연습문제 8.2: 1부터 1,000,000까지 정수의 합을 구하여 반환하는 함수 sum1to1000000()을 작성하여라.

```python
def sum1to1000000():
    sum = 0
    for i in range(1, 1000001):
        sum += i
    return sum

print(sum1to1000000())
```

```python
# (1) 이 계산을 수행하는 데 걸리는 시간을 초 단위로 구하여 출력하여라.
import time

start_time = time.time()
sum1to1000000()
end_time = time.time()

gap = end_time - start_time
print(f"1에서 1,000,000까지의 합을 구하는 시간 : {gap:.4f}초")
```

```python
# (2) 이 함수를 100번 호출하여 다음과 같이 소요된 시간을 출력하여라.
start_time = time.time()
for i in range(100):
    sum1to1000000()
end_time = time.time()

gap = end_time - start_time
print(f"1,000,000까지의 합을 100번 반복해서 구하는 시간 : {gap:.4f}초")
```

#### 연습문제 8.3: 이전 문제에서 사용한 함수 대신 다음과 같은 함수를 100번 수행하는 데 소요되는 시간을 출력하여라.

```python
# (1) 'Hello Python!'을 출력하는 함수
import time

def print_hello():
    print("Hello Python!")

start_time = time.time()
for i in range(100):
    print_hello()
end_time = time.time()

gap = end_time - start_time
print(f"Hello Python! 출력을 100번 반복하는 시간 : {gap:.4f}초")
```

```python
# (2) 10000 팩토리얼을 구하는 함수
import time

def factorial_10000():
    factorial = 1
    for i in range(1, 10001):
        factorial *= i
    return factorial

start_time = time.time()
for i in range(100):
    factorial_10000()
end_time = time.time()

gap = end_time - start_time
print(f"10000!을 100번 반복해서 구하는 시간 : {gap:.4f}초")
```

```python
# (3) 1에서 10,000까지의 모든 홀수를 세제곱하여 더한 값을 반환하는 함수
import time

def sum():
    sum = 0
    for i in range(1, 10001, 2):
        sum += (i**3)
    return sum

start_time = time.time()
for i in range(100):
    sum()
end_time = time.time()

gap = end_time - start_time
print(f"1에서 10000까지 홀수의 세제곱 더하기를 100번 반복하는 시간 : {gap:.4f}초")
```

```python
# (4) 1도에서 360도 사이의 sin 값을 1도 단위로 구하여 그 합을 반환하는 함수
import time
import math

def sin_sum():
    total = 0
    for i in range(1, 361):
        total += math.sin(math.radians(i))
    return total

start_time = time.time()
for i in range(100):
    sin_sum()
end_time = time.time()

gap = end_time - start_time
print(f"1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : {gap:.4f}초")
```

#### 연습문제 8.3: 0부터 1,000,000까지의 임의의 정수를 다음과 같이 연속적으로 10개 출력하는 의사난수 함수 my_rand()를 생성하여라.

```python
# (1) 이 의사난수를 만들기 위한 계수 값과 식을 적으시오.
def my_rand(x):
    # Numerical Recipes에서 유명한 LCG 계수
    a = 1664525
    b = 1013904223
    
    m = 1000001 # 생성된 난수가 0 ~ 1000000 사이어야 하므로
    
    next_x = (a * x + b) % m
    return next_x

x = 42
for _ in range(10):
    x = my_rand(x)
    print(x)
```

```python
# (2)-1 이 의사난수를 만들기 위한 초기 시드 값으로 time 모듈의 time() 함수를 사용하여 다시 코딩하시오.
#       (힌트: time() 값은 실수 값이므로 이를 정수로 변환하여 사용해야 한다)
import time

def my_rand(x):
    # Numerical Recipes에서 유명한 LCG 계수
    a = 1664525
    b = 1013904223
    
    m = 1000001 # 생성된 난수가 0 ~ 1000000 사이어야 하므로
    
    next_x = (a * x + b) % m
    return next_x

x = int(time.time())
for _ in range(10):
    x = my_rand(x)
    print(x)
```

```python
# (2)-2 이번에는 0에서 1,000,000까지의 임의의 5개의 정수를 출력하고 이를 다음과 같이 2번 반복하여라.
x = int(time.time())

print("1st:")
for _ in range(5):
    x = my_rand(x)
    print(x)

print("\n2nd:")
for _ in range(5):
    x = my_rand(x)
    print(x)
```
